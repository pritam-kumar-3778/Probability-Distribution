```python
# A fair coin is toss 5 times. Find the probablity that exactly 3 heads are obtained.
```

```python
from scipy.stats import binom

n = 5 # How many times event/trail occur
k = 3 # Success rate
p = 0.5 # Probablity of successer

probablity = binom.pmf(k, n, p) # pmf --> Probablity mass function
print(probablity) # 5/16
```
